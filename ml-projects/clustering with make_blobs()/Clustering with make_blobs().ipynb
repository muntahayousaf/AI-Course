# Clustering with make_blobs()

### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
```

### Generate Synthetic Dataset

```python
X, y_true = make_blobs(
    n_samples=300,        
    centers=3,            
    cluster_std=1.0,     
    random_state=42
)

print("Shape of features (X):", X.shape)
print("Shape of labels (y):", y_true.shape)

# Visualize the generated blobs
plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], c=y_true, cmap='viridis', s=40, edgecolor='k')
plt.title("Generated Gaussian Blobs (Ground Truth)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

### Apply KMeans Clustering

```python
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Visualize the clustering results
plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], c=y_kmeans, cmap='plasma', s=40, edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:,0], 
            kmeans.cluster_centers_[:,1], 
            s=200, c='red', marker='X', label='Centroids')
plt.title("KMeans Clustering on make_blobs() Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

### Compare True vs Predicted Labels

```python
score = adjusted_rand_score(y_true, y_kmeans)
print(" Adjusted Rand Index:", score)
```
